```python
from nltk import ngrams

sentence = "a bump on the log in the hole in the bottom of the sea"
sentence2 = "a frog on the bump on the log in the hole in the bottom of the sea"

# sentence = "your mother drives you in the car"
# sentence2 = "in mother russia car drives you"

n = 2
alphagrams = ngrams(sentence.split(), n)
betagrams = ngrams(sentence2.split(), n)
mygrams = list(alphagrams)
yourgrams = list(betagrams)
intersectiongrams=[]
uniongrams=[]

for i in range(len(mygrams)):
    for j in range(len(yourgrams)):
        if mygrams[i] == yourgrams[j]:
            if mygrams[i] not in intersectiongrams:
                intersectiongrams.append(mygrams[i])
   
for i in range(len(mygrams)):
    for j in range(len(yourgrams)):
            if mygrams[i] not in uniongrams:
                #print(mygrams[i])
                uniongrams.append(mygrams[i])
            if yourgrams[j] not in uniongrams:
                #print(yourgrams[j])
                uniongrams.append(yourgrams[j])
# print(mygrams)
# print('/')
# print(yourgrams)

print(intersectiongrams)
print(uniongrams)

similarity = len(intersectiongrams)/len(uniongrams)

print(round(similarity,2))

            
# for grams in sixgrams:
#   print(grams)
```

```python
def term_frequency(word_bag, corpus):
    a = {}
    word_bag=list(word_bag)
    for index in range(len(corpus)):
        v = []
        for i in range (len(word_bag)):
            if word_bag[i] in corpus[index]:
                value = corpus[index].count(word_bag[i])
                v.append(value)
            else:
                v.append(0)
        a[index] = v
        
    dataframe=pd.DataFrame(a, index=word_bag)
    dataframe=dataframe.rename(columns={0: 'Abstract',1: 'Introduction',2: 'Title'})
    return dataframe
    print(dataframe)
```

```python
dfD1 = term_frequency(word_bagD1,corpusD1)
resultsD1 = dfD1.transpose()

dfD2 = term_frequency(word_bagD2,corpusD2)
resultsD2 = dfD2.transpose()

dfD3 = term_frequency(word_bagD3,corpusD3)
resultsD3 = dfD3.transpose()

dfD4 = term_frequency(word_bagD4,corpusD4)
resultsD4 = dfD4.transpose()

dfD5 = term_frequency(word_bagD5,corpusD5)
resultsD5 = dfD5.transpose()
```
